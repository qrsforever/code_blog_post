```python
# 抽样分布

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
from scipy.stats import t
from scipy.stats import f
from scipy.stats import norm
```

```python
%matplotlib inline

plt.rcParams['figure.figsize'] = (6.0, 6.0)
```

## Pearson的$\chi^2$-分布

```python
chi2_froz_1 = chi2(5)
chi2_froz_2 = chi2(15)
chi2_froz_3 = chi2(35)

xs_nd = np.linspace(3, 50, num=200)

chi2_ys_nd_1 = chi2_froz_1.pdf(xs_nd)
chi2_ys_nd_2 = chi2_froz_2.pdf(xs_nd)
chi2_ys_nd_3 = chi2_froz_3.pdf(xs_nd)

print(chi2_froz_1.stats(moments="mvsk"))
print(chi2_froz_2.stats(moments="mvsk"))
print(chi2_froz_3.stats(moments="mvsk"))
```

```python
plt.plot(xs_nd, chi2_ys_nd_1, color='r', lw=2, label='df = 5')
plt.plot(xs_nd, chi2_ys_nd_2, color='g', lw=2, label='df = 15')
plt.plot(xs_nd, chi2_ys_nd_3, color='b', lw=2, label='df = 35')
plt.legend()
```

## Gosset的t-分布

```python
t_froz_1 = t(1)
t_froz_2 = t(3)
t_froz_3 = t(16)

xs_nd = np.linspace(-4, 4, num=20)
norm_ys_nd = norm.pdf(xs_nd)

t_ys_nd_1 = t_froz_1.pdf(xs_nd)
t_ys_nd_2 = t_froz_2.pdf(xs_nd)
t_ys_nd_3 = t_froz_3.pdf(xs_nd)

print(t_froz_1.stats(moments="mvsk"))
print(t_froz_2.stats(moments="mvsk"))
print(t_froz_3.stats(moments="mvsk"))
```

```python
plt.plot(xs_nd, t_ys_nd_1, color='r', lw=2, label='df = 1')
plt.plot(xs_nd, t_ys_nd_2, color='g', lw=2, label='df = 3')
plt.plot(xs_nd, t_ys_nd_3, color='b', lw=2, label='df = 10')

# 自由度越大, 月接近正态分布
plt.plot(xs_nd, norm_ys_nd, color='black', lw=2, label='norm')
plt.legend()
```

## Fisher的$\mathcal{F}$-分布

```python
f_froz_1 = f(1, 1)
f_froz_2 = f(15, 12)
f_froz_3 = f(100, 100)

xs_nd = np.linspace(f_froz_3.ppf(0.01), f_froz_3.ppf(0.99), 100)

f_ys_nd_1 = f_froz_1.pdf(xs_nd)
f_ys_nd_2 = f_froz_2.pdf(xs_nd)
f_ys_nd_3 = f_froz_3.pdf(xs_nd)

print(f_froz_1.stats(moments="mvsk"))
print(f_froz_2.stats(moments="mvsk"))
print(f_froz_3.stats(moments="mvsk"))

```

```python
plt.plot(xs_nd, f_ys_nd_1, color='r', lw=2, label='df1 = 1, df2 = 1')
plt.plot(xs_nd, f_ys_nd_2, color='g', lw=2, label='df1 = 15, df2 = 12')
plt.plot(xs_nd, f_ys_nd_3, color='b', lw=2, label='df1 = 100, df2 = 100')
plt.legend()
```
