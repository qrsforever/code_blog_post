```python
import numpy as np
import matplotlib.pylab as plt
```

```python
%matplotlib inline
```

```python
# 每股收益(每股利润), EPS:earnings_per_share
# 每股净资产, NA:net_assets_per_share
# 净资产收益率ROE(Rate of Return on Common Stockholders' Equity)股东权益报酬率
# 营业收入(Operating income)
# 净利润(Net Profit)

# EPS, NA, ROE, OI, NP 
labels = np.array(['每股收益', '每股净资产', '净资产收益率', '营业收入', '净利润'])
boe_data = np.array([0.1, 2.47, 4.0, 971.09, 34.35])
avg_data = np.array([0.4758, 4.6363, 10.63, 108.07, 4.49])
```

```python
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# 闭合
angles = np.concatenate((angles, [angles[0]]))

boe_data = np.concatenate((boe_data, [boe_data[0]]))
avg_data = np.concatenate((avg_data, [avg_data[0]]))
```

```python
fig = plt.figure()
# 极坐标
ax = fig.add_subplot(111, polar=True)
# 坐标起始方向
ax.set_theta_zero_location('N')
# 不显示极坐标最外圈的圆
# ax.spines['polar'].set_visible(False) 
# 不显示默认的分割线
# ax.grid(False)
# 不显示坐标间隔
# ax.set_yticks([]) 
ax.plot(angles, boe_data, 'ro-', linewidth=2)
ax.fill(angles, boe_data, facecolor='r', alpha=0.25)
ax.plot(angles, avg_data, 'bo-', linewidth=2)
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
ax.set_title("", va='bottom', fontproperties="SimHei")
ax.set_rlim(0, 10)
```
